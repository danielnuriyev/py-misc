```python
import pandas as pd
```

```python
# ss_events

sensor_counts_df = pd.read_csv('sensor_counts.csv')

sensor_counts_df = pd.pivot_table(
 sensor_counts_df, 
 index="uid", 
 columns="sensor_type" , 
 values="c", 
 fill_value=0.0).reset_index()

for c in sensor_counts_df.columns:
 if c == "uid": continue
 # sensor_counts_df[c] = sensor_counts_df[c].apply(lambda x: True if x > 0 else False) 

sensor_counts_df.head(5)
```

```python
# ss_events

event_type_counts_df = pd.read_csv('event_type_counts.csv')

event_type_counts_df = pd.pivot_table(
 event_type_counts_df, 
 index="uid", 
 columns="event_type" , 
 values="c", 
 fill_value=0.0).reset_index()

for c in event_type_counts_df.columns:
 if c == "uid": continue
 # event_type_counts_df[c] = event_type_counts_df[c].apply(lambda x: True if x > 0 else False)

event_type_counts_df.head(5)
```

```python
# sdl_customer_segmentation_usage_daily

event_type_counts_2_df = pd.read_csv('event_type_counts_2.csv')

col_names = []
for c in event_type_counts_2_df.columns:
 if c == "uid" or c.endswith("_event"):
    col_names.append(c)

event_type_counts_2_df = event_type_counts_2_df[col_names].fillna(0.0)

print(event_type_counts_2_df.shape)
event_type_counts_2_df.head(5)
```

```python
# sdl_app_events

device_counts_df = pd.read_csv("device_counts.csv").fillna(0.0)

device_counts_df.head(5)
```

```python
# sdl_sa_ss_location_audit

family_df = pd.read_csv("family.csv")
family_df = family_df[family_df.adults > 0]
print(family_df.shape)
family_df.head(5)
```

```python
subs_df = pd.read_csv('subs.csv')
# subs_df["has_prior_subscriptions"] = subs_df["has_prior_subscriptions"].apply(lambda x: 1.0 if x else 0.0)
subs_df.head(5)
```

```python
print(subs_df.shape)
print(subs_df["uid"].unique().shape)
print(sensor_counts_df.shape)
print(sensor_counts_df["uid"].unique().shape)
print(event_type_counts_df.shape)
print(event_type_counts_df["uid"].unique().shape)
print(event_type_counts_2_df.shape)
print(event_type_counts_2_df["uid"].unique().shape)
print(device_counts_df.shape)
print(device_counts_df["uid"].unique().shape)
print(family_df.shape)
print(family_df["uid"].unique().shape)
```

```python
df = pd.merge(subs_df, event_type_counts_df, on="uid", how="left").fillna(0.0)
df = pd.merge(df, sensor_counts_df, on="uid", how="left").fillna(0.0)
df = pd.merge(df, event_type_counts_2_df, on="uid", how="left").fillna(0.0)
df = pd.merge(df, family_df, on="uid", how="left").fillna(0.0)
df = pd.merge(df, device_counts_df, on="uid", how="left").fillna(0.0)

df.columns = df.columns.astype(str)
df.head(5)
```

```python
tmp = df[["subscription_status", "days"]].copy()
tmp.groupby(["days", "subscription_status"]).size()
```

```python
sub_df = df #.sample(frac=0.25)

y = sub_df["subscription_status"]
X = sub_df.drop("subscription_status", axis=1)
# X = X.drop(["has_prior_subscriptions","doorbell_count","residence","58","doorlock_count"], axis=1)
# tmp = X.nunique().sort_values(ascending=False)
# tmp.quantile(0.67)

X.head(5)
```

```python

tmp = tmp[tmp <= 8.4].index
X = X[tmp]
X.head(5)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True)
X_train.head(5)
```

```python
from sklearn.feature_selection import SelectPercentile, SelectKBest
from sklearn.feature_selection import f_classif

# fs = SelectPercentile(
#  score_func=f_classif
# )

fs = SelectKBest(score_func=f_classif, k=7)

fs.fit(X_train, y_train)

cols_idxs = fs.get_support(indices=True)

X_train_sub = X_train.iloc[:,cols_idxs]
X_test_sub = X_test.iloc[:,cols_idxs]

X_train_sub.head(5)
```

```python
def srt(x):
 return fs.scores_[x]

cols_idxs = sorted(cols_idxs, key=srt, reverse=True)

for i in cols_idxs:
 print(f"{X_train.columns[i]} : {fs.scores_[i]} : {fs.pvalues_[i]}")
```

```python
from matplotlib import pyplot

pyplot.bar([i for i in range(len(cols_idxs))], [fs.scores_[x] for x in cols_idxs])
pyplot.show()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import (
    accuracy_score,
    f1_score,
)
from catboost import CatBoostClassifier

model = GaussianNB()
# model = KNeighborsClassifier()
# model = svm.SVC()
# model = SGDClassifier()
# model = CatBoostClassifier()

# Model training
model.fit(X_train_sub, y_train)

# Predict Output
y_pred = model.predict(X_test_sub)

accuracy = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuracy)
print("F1 Score:", f1)
```
